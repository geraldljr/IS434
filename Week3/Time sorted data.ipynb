```python
import pandas as pd
df = pd.read_csv('firstTest.csv')
df.head()
```

```python
import numpy as np
import re

stopwords = ["for", "on", "an", "a", "of", "and", "in", "the", "to", "from",'see','will','market','sushi']

def clean_tweet(tweet):
    if type(tweet) == np.float:
        return ""
    temp = tweet.lower()
    temp = re.sub("'", "", temp) # to avoid removing contractions in english
    temp = re.sub("@[A-Za-z0-9_]+","", temp)
    temp = re.sub("#[A-Za-z0-9_]+","", temp)
    temp = re.sub(r'http\S+', '', temp)
    temp = re.sub('[()!?]', ' ', temp)
    temp = re.sub('\[.*?\]',' ', temp)
    temp = re.sub("[^a-z0-9]"," ", temp)
    temp = temp.split()
    temp = [w for w in temp if not w in stopwords]
    temp = " ".join(word for word in temp)
    return temp
```

```python
by_timestamp = {}
for row in df.iterrows():
    by_month = row[1]['timestamp'].split('T')[0].split('-')[0:2]
    to_add = ''.join(by_month)
    if to_add not in by_timestamp:
        cleaned = clean_tweet(row[1]['content'])
        by_timestamp[to_add] = [cleaned]
    else:
        cleaned = clean_tweet(row[1]['content'])
        by_timestamp[to_add].append(cleaned)
res = sorted(by_timestamp.items()) 
#format: [('yearmonth',[list of contents]),('yearmonth',[list of contents]),('yearmonth',[list of contents])]
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

for tup in res:
    print('for month',str(tup[0]))
    comment_words = ''
    # iterate through list
    for tweet_text in tup[1]:
        # split the text into tokens
        tokens = tweet_text.split()
        # Append tokens to string comment_words
        comment_words += " ".join(tokens)+" "

    # Let's make a word cloud
    wordcloud = WordCloud(width = 800, height = 800,
                    background_color ='white',
                    #stopwords = stopwords,
                    min_font_size = 10).generate(comment_words)

    # plot the WordCloud image                      
    plt.figure(figsize = (8, 8), facecolor = None)
    plt.imshow(wordcloud)
    plt.axis("off")
    plt.tight_layout(pad = 0)

    plt.show()
```
